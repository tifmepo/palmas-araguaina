```python
import pandas as pd
from openpyxl import load_workbook
import plotly.offline as py
import plotly.graph_objs as go
import plotly.express as px

py.init_notebook_mode(connected=True)

wb = load_workbook('corona.xlsx')
sheet = wb.active
data = []
palmas = []
araguaina = []
palmas1 = []
araguaina1 = []

for value,value2,value3,value4,value5 in sheet.iter_rows(min_row=2,
                             min_col=1,
                             max_col=5,
                             values_only=True):
    #print(value)
    
    data.append(value)
    palmas.append(value2)
    araguaina.append(value3)
    palmas1.append(value4)
    araguaina1.append(value5)
    
    
layout = go.Layout(title={
                        'text': "Palmas x Araguaina<BR>Casos Confirmados* - COVID-19",
                        'y':0.9,
                        'x':0.5,
                        'xanchor': 'center',
                        'yanchor': 'top'},
                   yaxis={'title':'Casos Confirmados'},
                   xaxis={'title':'*Dados retirados dos boletins diários das Prefeituras, diverge dos dados oficiais do Governo Estadual.'},
                   legend_title_text='Cidades<BR>'
                   )

trace = go.Scatter(x = data,
                   y = palmas,
                   mode = 'markers+lines',
                   name = 'Palmas',
                   #text=palmas,
                   textposition="top center")
trace1 = go.Scatter(x = data,
                   y = araguaina,
                   mode = 'markers+lines',
                   name = 'Araguaina',
                   #text=araguaina,
                   textposition="top center")

trace2 = go.Scatter(x = data,
                   y = palmas1,
                   mode = 'markers+lines',
                   name = 'Palmas-crescimento',
                   #text=palmas,
                   textposition="top center",
                   line=dict(dash='dot'))
trace3 = go.Scatter(x = data,
                   y = araguaina1,
                   mode = 'markers+lines',
                   name = 'Araguaina-crescimento',
                   #text=araguaina,
                   textposition="top center",
                   line=dict(dash='dot'))





data = [trace,trace1,trace2,trace3]

# Criando figura que será exibida

fig = go.Figure(data=data, layout=layout)

fig.update_xaxes(
    #rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=1, label="YTD", step="year", stepmode="todate"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(step="all")
        ])
    )
)



fig.update_yaxes(zeroline=True, zerolinewidth=2, zerolinecolor='black')


i=-1
while araguaina[i] is None :
    i=i-1

maiora = araguaina[i]

i=-1
while palmas[i] is None :
    i=i-1

maiorp = palmas[i]

if maiora >= maiorp :
    maior = maiora
else:
    maior = maiorp

fig.update_yaxes(range=[0,maior+20])
fig.update_xaxes(tickangle=-45)

fig.update_layout(hovermode="x")


# Exibindo figura/gráfico


fig.write_html("grafico.html")
py.iplot(fig)
```
